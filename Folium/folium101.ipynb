**Adding markers**

```python
m = folium.Map([41.0082, 28.97842], zoom_start=12, tiles="OpenStreetMap ")

folium.Marker(
    location=[41.142906, 29.049052],
    tooltip="Click me!",
    popup="Buluttan Weather Intelligence ",
    icon=folium.Icon(icon="cloud"),
).add_to(m)

folium.Marker(
    location=[41.0082, 28.9784],
    tooltip="Click me!",
    popup="Ayasofya",
    icon=folium.Icon(color="red"),
).add_to(m)

m
```

**Circle(),CircleMarker() ve Marker()**

```python
m = folium.Map(location=[41,29],tiles="Stamen Toner",
width="%100",
height="%100",
zoom_start=10)
folium.Circle(
radius=5000,
location=[41,29],
color='crimson',
fill=False,).add_to(m)
m
```

```python
# Another way to do this is to use the CircleMarker() function


m = folium.Map(location=[41,29],tiles="Stamen Toner",width="%100",height="%100")
folium.CircleMarker(location=(41,29),radius=100, fill_color='red').add_to(m)
m
```

```python
m = folium.Map(location=[40.75834,30.38001])
folium.Marker(location=[40.75834,30.38001], popup =  'Sakarya').add_to(m)
folium.CircleMarker(location=(40.75834,30.38001),radius=100, fill_color='blue').add_to(m)
m
```

**MiniMap()**

```python
from folium import plugins
m = folium.Map(location=(42, 29), zoom_start=5)
minimap = plugins.MiniMap()
m.add_child(minimap)
m
```

**GeoJSON File**

```python
from folium import GeoJson
geo=r"/content/cities.json"
file = open(geo, encoding="utf8")
text = file.read()
m = folium.Map(width="%100",weight="%100")
GeoJson(text).add_to(m)
m
```

```python
m = folium.Map(tiles="OpenStreetMap")
GeoJson(text,
       style_function=lambda feature: {
        'fillColor': '#adff2f',
        'color':None,
        'weight': 1,
        'dashArray': '4, 4'
    }).add_to(m)
m
```

```python
m = folium.Map(tiles="Cartodbdark_matter",width="%100",weight="%100")
GeoJson(text,).add_to(m)
m
```

```python
import requests

m = folium.Map(tiles="cartodbpositron")

geojson_data = requests.get(
    "https://raw.githubusercontent.com/python-visualization/folium-example-data/main/world_countries.json"
).json()

folium.GeoJson(geojson_data, name="hello world").add_to(m)

folium.LayerControl().add_to(m)

m
```
